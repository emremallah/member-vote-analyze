## For numeric columns: What are the max, min, mean, and median?

```python
s = data.describe()
```

```python
for i in s.columns:
  print(f"{i} : {data[i].max()}")
  print(f"{i} : {data[i].min()}")
  print(f"{i} : {data[i].mean()}")
  print(f"{i} : {data[i].median()}")
```

## For text columns: What is the most common value? How many unique values are there?

```python
d = data.drop("_id" , axis=1)
for i in d.columns:
  print(data[i].value_counts().head(1))
  # print(data[i].unique())
  print(f"{i} : {data[i].nunique()} \n")
```

## Plot two or more columns in your data using matplotlib, seaborn, or plotly. Make sure that your plot has labels, a title, a grid, and a legend.

```python
import matplotlib.pyplot as plt
x = data[data["Vote"] == "Yes"]
y = data[data["Vote"] == "No"]
x1 = ["YES" , "NO"]
y1 = [len(x) , len(y)]
plt.bar(x1 , y1)
plt.title("PLOTING")
plt.xlabel("Y/N")
plt.ylabel("LENGTH OF YES AND NO")
plt.grid()
plt.legend()
plt.show()
```
